## 2. Chargement et préparation des données

Les données sont extraites du rapport Stanford HAI AI Index 2025 (Chapitre 4 — Economy) et de Statista. Elles couvrent la période 2015–2025 pour les investissements privés mondiaux en IA.

## 3. Analyse Exploratoire des Données (EDA)

Avant de visualiser, on calcule les statistiques clés pour comprendre la structure des données.

```python
print('=' * 55)
print('  STATISTIQUES CLÉS — INVESTISSEMENTS IA MONDIAUX')
print('=' * 55)

# Calcul du TCAM (Taux de Croissance Annuel Moyen)
debut = df_year[df_year.year == 2015]['world'].values[0]
fin   = df_year[df_year.year == 2024]['world'].values[0]
n     = 2024 - 2015
tcam  = (fin / debut) ** (1/n) - 1

debut_us = df_year[df_year.year == 2015]['usa'].values[0]
fin_us   = df_year[df_year.year == 2024]['usa'].values[0]
tcam_us  = (fin_us / debut_us) ** (1/n) - 1

# Part des États-Unis en 2024
world_2024 = df_year[df_year.year == 2024]['world'].values[0]
usa_2024   = df_year[df_year.year == 2024]['usa'].values[0]
china_2024 = df_year[df_year.year == 2024]['china'].values[0]
part_us    = usa_2024 / world_2024 * 100
ratio_us_cn = usa_2024 / china_2024

# Croissance 2024 -> 2025
world_2025 = df_year[df_year.year == 2025]['world'].values[0]
croiss_2025 = (world_2025 - world_2024) / world_2024 * 100

print(f'\n  Investissement mondial 2024 : ${world_2024:.0f} Md')
print(f'  Investissement mondial 2025 : ${world_2025:.0f} Md')
print(f'  Croissance 2024→2025        : +{croiss_2025:.0f}%')
print(f'\n  TCAM mondial (2015–2024)    : {tcam*100:.1f}% / an')
print(f'  TCAM États-Unis (2015–2024) : {tcam_us*100:.1f}% / an')
print(f'\n  Part des États-Unis (2024)  : {part_us:.0f}%')
print(f'  Ratio US / Chine (2024)     : {ratio_us_cn:.0f}x')
print(f'\n  Top secteur 2024            : Infrastructure IA ($80Md)')
print(f'  IA Générative 2024          : $33.9Md (+18.7% vs 2023)')
print('=' * 55)

print('\nAperçu du dataset temporel :')
df_year
```

```python
print('Classement des pays par investissement 2024 :\n')
df_rank = df_country.sort_values('investment_2024_bn', ascending=False).reset_index(drop=True)
df_rank.index += 1
df_rank['Part mondiale (%)'] = (df_rank['investment_2024_bn'] / df_rank['investment_2024_bn'].sum() * 100).round(1)
print(df_rank[['country','investment_2024_bn','Part mondiale (%)']].to_string())
```

## 4. Visualisations

### 4.1 — Évolution temporelle des investissements (2015–2025)

**Lecture :** Ce graphique montre la progression des investissements privés en IA sur 11 ans. On observe deux phases distinctes : une croissance modérée jusqu'en 2020, puis une explosion à partir de 2021 avec l'essor de l'IA générative.

```python
fig, ax = plt.subplots(figsize=(14, 7))
fig.patch.set_facecolor('#0F1117')
ax.set_facecolor('#0F1117')

# Courbes
ax.plot(df_year['year'], df_year['world'], 'o-',
        color='#378ADD', linewidth=3, markersize=7, label='Monde', zorder=3)
ax.plot(df_year['year'], df_year['usa'], 's--',
        color='#1D9E75', linewidth=2.5, markersize=6, label='États-Unis', zorder=3)
ax.plot(df_year['year'], df_year['china'], '^:',
        color='#D85A30', linewidth=2.5, markersize=6, label='Chine', zorder=3)

# Remplissage
ax.fill_between(df_year['year'], df_year['world'], alpha=0.1, color='#378ADD')
ax.fill_between(df_year['year'], df_year['usa'],   alpha=0.07, color='#1D9E75')

# Annotations
ax.annotate('Boom IA générative\n+132% en un an\n$93Md',
            xy=(2021, 93), xytext=(2018.5, 170),
            arrowprops=dict(arrowstyle='->', color='#aaa', lw=1.2),
            fontsize=10, color='#aaa', ha='center')
ax.annotate('ChatGPT (nov. 2022)\ndéclenche une vague\nd\'investissements',
            xy=(2023, 95), xytext=(2020.5, 300),
            arrowprops=dict(arrowstyle='->', color='#aaa', lw=1.2),
            fontsize=10, color='#aaa', ha='center')
ax.annotate('Record absolu 2025\n$581Md (+130%)',
            xy=(2025, 581), xytext=(2023.2, 490),
            arrowprops=dict(arrowstyle='->', color='#378ADD', lw=1.5),
            fontsize=11, color='#378ADD', fontweight='bold')

# Axes
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'${x:.0f}Md'))
ax.set_xticks(df_year['year'])
ax.set_xlim(2014.5, 2025.5)

# Titres
ax.set_title('Où va l\'argent dans l\'IA ? — Évolution 2015–2025',
             fontsize=17, fontweight='bold', color='white', pad=20)
ax.set_xlabel('Année', labelpad=10)
ax.set_ylabel('Investissements privés (Milliards USD)', labelpad=10)

legend = ax.legend(frameon=True, facecolor='#1a1a2e',
                   edgecolor='#333', labelcolor='white', fontsize=12,
                   loc='upper left')

plt.tight_layout()
plt.savefig('graphique_1_evolution.png', dpi=150, bbox_inches='tight',
            facecolor='#0F1117')
plt.show()
print('Graphique sauvegardé : graphique_1_evolution.png')
```

**Remarque :** On distingue clairement deux phases :
- **Phase 1 (2015–2020)** : Croissance lente et régulière, +27% par an en moyenne.
- **Phase 2 (2021–2025)** : Explosion exponentielle, portée par l'IA générative et les méga-investissements (OpenAI $6.6Md, xAI $6Md, Anthropic $4Md en 2024).

L'écart US / Chine s'est creusé : de 2x en 2015 à **23x en 2025**.

### 4.2 — Top pays par investissement privé en IA (2024)

```python
df_c = df_country.sort_values('investment_2024_bn')
colors = ['#378ADD' if c == 'États-Unis' else
          '#1D9E75' if c in ['Royaume-Uni', 'France', 'Allemagne', 'Suède'] else
          '#D85A30' if c == 'Chine' else
          '#7F77DD' for c in df_c['country']]

fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('#0F1117')
ax.set_facecolor('#0F1117')

bars = ax.barh(df_c['country'], df_c['investment_2024_bn'],
               color=colors, height=0.65, zorder=2)

for bar, val in zip(bars, df_c['investment_2024_bn']):
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
            f'${val:.1f}Md', va='center', color='white', fontsize=10)

ax.set_xlim(0, 135)
ax.set_title('Investissement privé en IA par pays · 2024',
             fontsize=16, fontweight='bold', color='white', pad=20)
ax.set_xlabel('Milliards USD', labelpad=10)

# Légende régions
from matplotlib.patches import Patch
legend_elems = [
    Patch(facecolor='#378ADD', label='Amérique du Nord (USA)'),
    Patch(facecolor='#D85A30', label='Chine'),
    Patch(facecolor='#1D9E75', label='Europe'),
    Patch(facecolor='#7F77DD', label='Reste du monde'),
]
ax.legend(handles=legend_elems, frameon=True, facecolor='#1a1a2e',
          edgecolor='#333', labelcolor='white', fontsize=10,
          loc='lower right')

plt.tight_layout()
plt.savefig('graphique_2_pays.png', dpi=150, bbox_inches='tight',
            facecolor='#0F1117')
plt.show()
print('Graphique sauvegardé : graphique_2_pays.png')
```

**Remarque :** Les États-Unis dominent de façon écrasante avec **$109.1Md** en 2024, soit plus de **11x** la Chine ($9.3Md). L'Europe (UK, Suède, France, Allemagne) représente collectivement environ $13Md, soit 12% des USA. Ce graphique confirme la concentration extrême de l'écosystème IA autour de la Silicon Valley.

## 6. Recommandations

Sur la base de cette analyse, voici les recommandations pour les différents acteurs :

### Pour les investisseurs
- **Privilégier l'infrastructure IA** (GPU, datacenters, cloud) : c'est la couche sur laquelle tout repose, avec des barrières à l'entrée élevées et des revenus récurrents.
- **Surveiller les marchés émergents** : les Émirats, l'Arabie Saoudite et l'Inde présentent des taux de croissance élevés avec des valorisations encore accessibles.
- **Diversifier vers la santé IA** : secteur en forte croissance (+62% H1 2025), soutenu par la réglementation (FDA) et des besoins structurels immenses.

### Pour les entreprises
- **Intégrer l'IA dans les processus métiers** dès maintenant : 78% des organisations l'utilisent déjà en 2024.
- **Former les équipes** aux outils IA : la pénurie de talents est le principal frein identifié.

### Pour les décideurs publics
- **Suivre l'exemple français** (€109Md) et saoudien ($100Md) : les gouvernements qui n'investissent pas massivement aujourd'hui risquent de perdre leur souveraineté technologique.
- **Développer des politiques de données** : la qualité des données nationales est le nouveau pétrole de l'IA.

## 7. Limites de l'étude

- Les données proviennent principalement du secteur **privé** — les investissements militaires et gouvernementaux chinois (estimés à $184Md entre 2000–2023) sont sous-représentés.
- Les chiffres 2025 sont partiellement des **projections** et estimations (source : Quantumrun, avril 2026).
- L'analyse ne couvre pas les **investissements en R&D interne** des grandes entreprises tech (Google, Microsoft, Meta, Apple) qui ne sont pas divulgués séparément.
